```python
import kagglehub
import pandas as pd
import os

# Download latest version
path = kagglehub.dataset_download("shivamb/netflix-shows")

csv_data = os.path.join(path,"netflix_titles.csv")

df = pd.read_csv(csv_data)
df.head()
```

**Data understanding**

```python
df.isnull().sum()
```

**Clenaing Data**

```python
# as see befor director have a lot of null value so we start by clenaing the esay one date_added rating duration
# There is not too much null value so the best option here is to delete the null row

df[df['date_added'].isnull()]
```

```python
df[df['rating'].isnull()]
```

```python
df.dropna(subset=["date_added", "rating", "duration"],inplace=True)
```

```python
df.isnull().sum()
```

```python
len(df[df['country'].isnull()])
```

```python
df["country"].value_counts().head(10)
```

```python
# as see we have 829 null country and country is import data so we can't delete so we going to replace all null country with Others

df['country'] = df['country'].fillna("Others")
```

```python
df["country"].value_counts().head(10)
```

```python
# also for director and cost we can't delete them beacuse we goning to lose about 2600 row for the data so we replace it with UNKNOW

df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
```
