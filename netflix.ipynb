```python
import kagglehub
import pandas as pd
import os

# Download latest version
path = kagglehub.dataset_download("shivamb/netflix-shows")

csv_data = os.path.join(path,"netflix_titles.csv")

df = pd.read_csv(csv_data)
df.head()
```

**Data understanding**

```python
df.shape

# we have 8807 row and 12 columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

**Clenaing Data**

```python
# as see befor director have a lot of null value so we start by clenaing the esay one date_added rating duration
# There is not too much null value so the best option here is to delete the null row

df[df['date_added'].isnull()]
```

```python
df[df['rating'].isnull()]
```

```python
df.dropna(subset=["date_added", "rating", "duration"],inplace=True)
```

```python
df.isnull().sum()
```

```python
len(df[df['country'].isnull()])
```

```python
df["country"].value_counts().head(10)
```

```python
# as see we have 829 null country and country is import data so we can't delete so we going to replace all null country with Others

df['country'] = df['country'].fillna("Others")
```

```python
df["country"].value_counts().head(10)
```

```python
# also for director and cost we can't delete them beacuse we goning to lose about 2600 row for the data so we replace it with UNKNOW

df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df['release_year']
```

```python
# for the release_year column i have see that there is only year not a date so i don't need to covered from integer to date_time
```

**Analatical part**

```python
import matplotlib.pyplot as plt
```

```python
# the first we show movies are more common than TV shows

df['type'].value_counts()
```

```python
df["type"].value_counts().plot(kind="bar")
plt.title("Movies vs TV Shows on Netflix")
plt.ylabel("Count")
plt.show()
```

We have 6126 movies and 2664 TV shows

```python
# now let see the years growth

df["release_year"].value_counts().sort_index().plot(figsize=(12,5))
plt.title("Netflix Content by Release Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

**Top 10 dominates country on Netflix content**

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries by Number of Titles")
plt.ylabel("Movies numbers")
plt.show()
```

**Which rating appears most often**

```python
df["rating"].value_counts().plot(kind="bar")
plt.title("Content Ratings Distribution")
plt.show()
```

```python
genres = df["listed_in"].str.split(", ")
```

**Top 10 Genres on Netfkix**

```python
genres.explode().value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Genres on Netflix")
plt.show()
```
